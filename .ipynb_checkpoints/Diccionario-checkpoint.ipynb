# Diccionarios

```python
# Crear Diccionario
dict = {'Nombre': 'Bryan',
        'Edad' : 19,
        'Altura' : 1.69
}
print(dict)

#for value in dict:
#    print(value)

# Recorrer Diccionario
for value in dict.keys():
    print(value)
for value in dict.values():
    print(value)

# Acceder a un elemento a travez de las keys en un diccionario
print(dict['Nombre'])

# Mostrar métodos asociados a la instancia
print(dir(dict))

# Mostrar lo que hacen los métodos de la instancia
print(help(dict))

# dict.items() Retorna una lista de tuplas [(key, value), (key, value), (key, value), ...]
print(dict.items())
for key, value in dict.items():
    print(key, value)

# Acceder a un elemento a travez de las keys en un diccionario
dict['Nombre'] = 'Freyra'
print(dict.items())
for key, value in dict.items():
    print(key, value)
```

```python
from math import sqrt

class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
        
    def total_seconds(self):
        return (self.h * 60 * 60)+(self.m * 60)+(self.s)

    def __str__(self):
        if(self.h < 10):
            self.h = f"0{self.h}"
        if(self.m < 10):
            self.m = f"0{self.m}"
        if(self.s < 10):
            self.s = f"0{self.s}"
        return f'{self.h}:{self.m}:{self.s}'
```

```python
now = Time(2,10,3)
print(now.total_seconds())
print(now)
```
